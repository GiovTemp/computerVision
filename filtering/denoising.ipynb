```python
# importing the modules needed
import cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
  
# Reading the image
image = mpimg.imread('noise.jpg')

#2d filtering
#kernel = np.array([[0.04, 0.04, 0.04, 0.04, 0.04],[0.04, 0.04, 0.04, 0.04, 0.04],[0.04, 0.04, 0.04, 0.04, 0.04],[0.04, 0.04, 0.04, 0.04, 0.04],[0.04, 0.04, 0.04, 0.04, 0.04]])
#kernel = np.array([[0.07, 0.07, 0.07, 0.07, 0.07],[0.07,0.07, 0.07,0.07, 0.07],[0.07,0.07, 0.07,0.07, 0.07],[0.07,0.07, 0.07,0.07, 0.07],[0.07,0.07, 0.07,0.07, 0.07]])
kernel = np.ones((5,5),np.float32)/25 

dst = cv2.filter2D(image,-1,kernel)

#Gaussian
imgGauss=cv2.GaussianBlur(image,(5,5),0)
#Blur
imgBlur=cv2.blur(image, ksize=(7,7))
#Median
imgMedian=cv2.medianBlur(image,7)
#Bilateral
bilateral = cv2.bilateralFilter(image,9,75,75)


images = [image,dst,imgGauss, imgBlur, imgMedian, bilateral]
fig, axs = plt.subplots(nrows = 1, ncols = 6, figsize = (25, 25))
for ind, p in enumerate(images):
    ax = axs[ind]
    ax.imshow(p)
    ax.axis('off')
axs[0].set_title('Original', fontsize=16)
axs[1].set_title('Low Pass Filters', fontsize=16)    
axs[2].set_title('Average blur', fontsize=16)
axs[3].set_title('Gaussian blur', fontsize=16)
axs[4].set_title('Median blur', fontsize=16)
axs[5].set_title('Bilateral blur', fontsize=16)
plt.show()



```

```python

# Reading the image
image = mpimg.imread('noise.webp')

#2d filtering
kernel = np.array([[0.04, 0.04, 0.04, 0.04, 0.04],[0.04, 0.04, 0.04, 0.04, 0.04],[0.04, 0.04, 0.04, 0.04, 0.04],[0.04, 0.04, 0.04, 0.04, 0.04],[0.04, 0.04, 0.04, 0.04, 0.04]])
#kernel = np.ones((5,5),np.float32)/25
dst = cv2.filter2D(image,-1,kernel)

#Gaussian
imgGauss=cv2.GaussianBlur(image, ksize=(5,5), sigmaX=0)
#Blur
imgBlur=cv2.blur(image, ksize=(7,7))
#Median
imgMedian=cv2.medianBlur(image,7)
#Bilateral
bilateral = cv2.bilateralFilter(image,9,75,75)
 

images = [image,dst,imgGauss, imgBlur, imgMedian,bilateral]
fig, axs = plt.subplots(nrows = 1, ncols = 6, figsize = (25, 25))
for ind, p in enumerate(images):
    ax = axs[ind]
    ax.imshow(p)
    ax.axis('off')
axs[0].set_title('Original', fontsize=16)
axs[1].set_title('Low Pass Filters', fontsize=16)    
axs[2].set_title('Average blur', fontsize=16)
axs[3].set_title('Gaussian blur', fontsize=16)
axs[4].set_title('Median blur', fontsize=16)
axs[5].set_title('Bilateral blur', fontsize=16)
plt.show()

```
