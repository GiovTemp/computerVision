```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

import matplotlib.image as mpimg

#read the image
img=mpimg.imread('dots.jpeg')
plt.imshow(img)
```

```python
# Copy the image
img_grid = img.copy()
w=int(img_grid.shape[1])
h=int(img_grid.shape[0])
grid_h = h/3
grid_w = w/3
# Draw the grid 
for i in range(1, 4):
    cv2.line(img_grid, (0,int(grid_h*i) ), (w, int(grid_h*i)), (0, 255, 0), 5)
    cv2.line(img_grid, (int(grid_w*i),0), (int(grid_w*i),h), (0, 255, 0), 5)
#last argument is thickness expressed in dots; 1 is the smallest value

plt.imshow(img_grid)
plt.show()
```

```python
# Copy the image
img_grid = img.copy()
center = (int(w/2),int(h/2))
# Draw 4 arrowed line
cv2.arrowedLine(img_grid, center, (w, 0), (0, 0, 0), 5)
cv2.arrowedLine(img_grid, center, (0, 0), (0, 0, 0), 5)
cv2.arrowedLine(img_grid, center, (0, h), (0, 0, 0), 5)
cv2.arrowedLine(img_grid, center, (h, w), (0, 0, 0), 5)
plt.imshow(img_grid)
plt.show()
```
