```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow

img = cv2.imread('flowers.jpg', -1)
image_rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB) #convert the image to BGR e RGB
plt.imshow(image_rgb)

cv2_imshow(img)
```

```python
scale=0.1
fig, axs = plt.subplots(nrows = 1, ncols = 3, figsize = (20, 20))
pixels = []
for i in range(0, 3):
    w=int(img.shape[1]*(scale + 0.2*i))
    h=int(img.shape[0]*(scale + 0.2*i))
    pixels.append(w*h)
    ax = axs[i]
    ax.imshow(cv2.resize(image_rgb,(w, h),interpolation = cv2.INTER_LINEAR))
axs[0].set_title('0.3 - tot pixels : '+str(pixels[0]))
axs[1].set_title('0.5 - tot pixels : '+str(pixels[1]))
axs[2].set_title('0.7  - tot pixels : '+str(pixels[2]))
plt.show()
```

```python
# Plot the three channels of the image
fig, axs = plt.subplots(nrows = 1, ncols = 3, figsize = (20, 20))
for i in range(0, 3):
    ax = axs[i]
    ax.imshow(image_rgb[:, :, i], cmap = 'gray')
axs[0].set_title('RED')
axs[1].set_title('GREEN')
axs[2].set_title('BLUE')
plt.show()
```
